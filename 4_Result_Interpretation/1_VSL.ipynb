## World bank gross national income (GNI) per capita 2015
$$
VSL_{c} = VSL_{US}\ \times\ (\frac{Y_{c}}{Y_{US}})^{\eta}
$$

```python
VSL_us = VSL_dict['USA']
Y_us = 55980
VSL_us * (7630/Y_us), VSL_dict['CHN']
```

```python
VSL_dict['NCL'] = VSL_us * (13210/Y_us) # New Caledonia; GNI: 13210 (2000)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.CD?locations=NC

VSL_dict['LIE'] = VSL_us * (116440/Y_us) # Liechtenstein; GNI: 116,440 (2009)
# Source:https://data.worldbank.org/indicator/NY.GNP.PCAP.CD?locations=LI

VSL_dict['LBY'] = VSL_us * (21570/Y_us) # Libya; GNI: 21570 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=LY

VSL_dict['TWN'] = VSL_us * (23367/Y_us) # Taiwan; GNI: 23367 (2015)
# Source: https://www.statista.com/statistics/321596/taiwan-per-capita-gross-national-product/

VSL_dict['PRK'] = VSL_us * (555/Y_us) # North Korea; GNI: 555 (2008)
# Source: https://www.nationsonline.org/oneworld/Country-Stats/Korea-North-statistics.htm

VSL_dict['GRL'] = VSL_dict['DNK'] # Greenland 
# Greenland is an island country that is part of the Kingdom of Denmark.

VSL_dict['MCO'] = VSL_us * ((6259130741/ population['MCO'])/Y_us) # Monaco; GNI: 6259130741 (2015 total)
# Source: http://data.un.org/Data.aspx?q=Monaco&d=SNAAMA&f=grID%3A103%3BcurrID%3AUSD%3BpcFlag%3A0%3BcrID%3A492

VSL_dict['ERI'] = VSL_us * (1610/Y_us)# ; GNI: 1610
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=ER

VSL_dict['VGB'] = VSL_dict['GBR'] # British Virgin Island
# British Virgin Islanders are British Overseas Territories citizens and since 2002, are also British citizens.


VSL_dict['ABW'] = VSL_us * (35740/Y_us)# ; GNI: 35740 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=AW

VSL_dict['CUB'] = VSL_us * (7230/Y_us)# ; GNI: 7230 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.CD?locations=CU

VSL_dict['PYF'] = VSL_us * (15770/Y_us)# ; GNI: 15770 (2000)
# Source: https://www.macrotrends.net/countries/PYF/french-polynesia/gni-per-capita

VSL_dict['DJI'] = VSL_us * (4510/Y_us)# ; GNI: 4510 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=DJ

VSL_dict['SMR'] = VSL_us * ((1208405538/population['SMR'])/Y_us) # San Marino; GNI: 1208405538 (2015 total)
# Source: http://data.un.org/Data.aspx?q=San+Marino&d=SNAAMA&f=grID%3A103%3BcurrID%3AUSD%3BpcFlag%3A0%3BcrID%3A674

VSL_dict['CYM'] = VSL_us * (38100/Y_us)# ; GNI: 38100 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=KY

VSL_dict['SYR'] = VSL_us * (1020/Y_us) # Syria; GNI: 1020 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.CD?locations=SY

VSL_dict['SOM'] = VSL_us * (1080/Y_us) # Somalia; GNI: 1080 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=SO

VSL_dict['VEN'] = VSL_us * (17090/Y_us) # Venezuela; GNI: 17090 (2015)
# Source: https://data.worldbank.org/indicator/NY.GNP.PCAP.PP.CD?locations=VE

VSL_dict['ROW'] = 0.107 # mean low income economy
```
