```python
# Row sum: total deaths happen in that country
con_total = dict(pd.DataFrame(con_mort_tol).sum(axis = 1))
pro_total = dict(pd.DataFrame(pro_mort_tol).sum(axis = 1))
```

```python
# Col sum: total deaths caused by that country
con_caused = dict(pd.DataFrame(con_mort_tol).sum(axis = 0))
pro_caused = dict(pd.DataFrame(pro_mort_tol).sum(axis = 0))
```

```python
# total deaths happen in that country
con_dom_world = {}
for country in con_mort_tol:
    con_dom_world[country] = {}
    con_dom_world[country]['domestic'] = con_mort_tol[country][country]
    con_dom_world[country]['world'] = con_total[country]
    con_dom_world[country]['OtherPartofWorld'] = con_dom_world[country]['world'] - con_dom_world[country]['domestic']
```

```python
# total deaths caused by that country
con_cause_dom_world = {}
for country in con_mort_tol:
    con_cause_dom_world[country] = {}
    con_cause_dom_world[country]['domestic'] = con_mort_tol[country][country]
    con_cause_dom_world[country]['world'] = con_caused[country]
    con_cause_dom_world[country]['OtherPartofWorld'] = con_cause_dom_world[country]['world'] - con_cause_dom_world[country]['domestic']
```

```python
net_export_death = {}
export_of_death = {}
import_of_death = {}

for country in con_cause_dom_world:
    if country != 'ROW':
        net_export_death[country] = (con_cause_dom_world[country]['OtherPartofWorld'] - con_dom_world[country]['OtherPartofWorld'])
        export_of_death[country] = con_cause_dom_world[country]['OtherPartofWorld']# total deaths caused by that country
        import_of_death[country] = con_dom_world[country]['OtherPartofWorld']# total deaths happen in that country
```

```python
Country_df_dict = {}
for country in Country_num_dict:
    Country_df_dict[country] = {}
    
    country_name = df_final[df_final['ADM0_A3']==country]['NAME'].values[0]
    continent = df_final[df_final['ADM0_A3']==country]['CONTINENT'].values[0]
    colony =  df_final[df_final['ADM0_A3']==country]['Colony'].values[0]
    num_suffer = Country_num_dict[country]
    Country_df_dict[country]['country_name'] = country_name
    Country_df_dict[country]['continent'] = continent
    Country_df_dict[country]['colony'] = colony
    Country_df_dict[country]['num_suffer'] = num_suffer
    
    # CONTINENT
    Country_df_dict[country]['num_suffer_Asia'] = [Country_continent_dict[country]['Asia'] if 'Asia' in Country_continent_dict[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_Africa'] = [Country_continent_dict[country]['Africa'] if 'Africa' in Country_continent_dict[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_Europe'] = [Country_continent_dict[country]['Europe'] if 'Europe' in Country_continent_dict[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_Oceania'] = [Country_continent_dict[country]['Oceania'] if 'Oceania' in Country_continent_dict[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_North_America'] = [Country_continent_dict[country]['North America'] if 'North America' in Country_continent_dict[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_South_America'] = [Country_continent_dict[country]['South America'] if 'South America' in Country_continent_dict[country].keys() else 0][0]

    # Colony
    Country_df_dict[country]['num_suffer_colony'] = [Country_colony_dict2[country]['Colony'] if 'Colony' in Country_colony_dict2[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_colonizer'] = [Country_colony_dict2[country]['Colonizer'] if 'Colonizer' in Country_colony_dict2[country].keys() else 0][0]
    Country_df_dict[country]['num_suffer_other'] = [Country_colony_dict2[country]['Other'] if 'Other' in Country_colony_dict2[country].keys() else 0][0]
#     print('Country Name:',country_name,' ','Continent:', continent, ' ',
#          'Colony status:', colony,' ','Number of suffering countries', num_suffer)
```
