```python
# Change 'PSX' to 'PSE' and change 'SDN' to  'SUD'
df_shp.loc[11,'ADM0_A3'] = 'PSE'
df_shp.loc[86,'ADM0_A3'] = 'SUD'
```

```python
# wholeworldB = gpd.read_file('output/output_wholeworld/wholeworld_bio.shp')
wholeworldB = gpd.read_file('C:/Users/15311/Desktop/output_wholeworld/wholeworld_bio.shp')
wholeworldB = gpd.GeoDataFrame(wholeworldB)
```

```python
wholeworldB['area_fraction'] = [{} for i in range(len(wholeworldB))]
```

```python
wholeworldB.iloc[1,13]
```

```python
for i in Country[:-1]:
    for j in range(len(wholeworldB)):
        a = df_shp[df_shp['ADM0_A3']==i].intersection(wholeworldB.iloc[j].geometry).area.values[0]
        if a == 0:
            pass
        else:
            wholeworldB.iloc[j,13][i] = a / wholeworldB.iloc[j].geometry.area
```

```python
# Summation of the value insde a dictionary
def dictsum(myDict):      
    sum = 0
    for i in myDict: 
        sum = sum + myDict[i] 
    return sum
```

```python
for i in range(len(wholeworldB)):
    if wholeworldB.iloc[i,13] == {}:
        wholeworldB.iloc[i,13]['ROW'] = 1
    elif dictsum(wholeworldB.iloc[i,13]) != 1:
        wholeworldB.iloc[i,13]['ROW'] = 1 - dictsum(wholeworldB.iloc[i,13])
```
